```python
import matplotlib.pyplot as plt

# 模型名称
models = ['Model 1', 'Model 2', 'Model 3']

# 每个模型在 real、AIGC 和 adv 上的数值（假设为一些示例数据）
real_values = [0.69, 0.86, 0.89]
aigc_values = [0.71, 0.83, 0.92]
adv_values = [0.43, 0.50, 0.75]

# 创建图形和轴
fig, ax = plt.subplots()

# 绘制每条折线
ax.plot(models, real_values, marker='o', label='Real')
ax.plot(models, aigc_values, marker='o', label='AIGC')
ax.plot(models, adv_values, marker='o', label='Adv')

ax.set_ylim(0, 1)

# 添加标题和标签
ax.set_title('Model Performance Comparison')
ax.set_xlabel('Models')
ax.set_ylabel('Values')

# 添加图例
ax.legend()

# 显示图形
plt.show()
```
